```python
### PLOTING POINTS AND MAKING GRAph###
import matplotlib.pyplot as plt
import numpy as np

 
xpoints = np.array([1,6])
ypoints=np.array([4,5])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
ypoints = np.array([2,5,7,10])

plt.plot(ypoints, '*''y')   # 'y'or'r' or 'k' represents the color of the symbol we are using
plt.show()
```

```python
xpoints=np.array([1,2,10,4])

plt.plot(xpoints, 'D''r')
plt.show()
```

```python
ypoints = np.array([2,5,4,9])
xpoints = np.array([1,6,8,3])
plt.plot(xpoints,ypoints, marker ='.' ,ms = 100 ,color = 'pink')   ## ms indicates MARKER SIZE(WIDTH)
plt.show()
```

```python
plt.plot(xpoints,ypoints,'*''g',ms ='100',mec = 'r')     ### mec stands for 'MARKER EDGE COLOR'
plt.show()
```

```python
plt.plot(xpoints,ypoints,marker= '*',ms ='30',mec = 'k',color= 'b')
plt.show()
```

```python
plt.plot(xpoints,ypoints,'*','g',ms ='50',mec = 'k') ## HERE IT WILL TAKE THE DEFAULT COLOR BCAUSE(,)
plt.show()
```

```python
plt.plot(xpoints,ypoints, marker ='.',mfc ='pink',mec = 'b', ms = '40', color = 'y')   ## mfc 'MARKER FACE COLOR'
plt.show()
```

```python
### USE BOTH MFC AND MEC ARGUMENTS TO COLOR OF THE ENTIRE MARKER
ypoint = np.array([3,2,6,5])
plt.plot(ypoints,marker = '*', ms = '60',mec = 'yellow' , mfc ='k')
plt.show()
```

```python
plt.plot(ypoints,marker = '*', linestyle = 'dashed',ms = '40',mec = 'k',mfc = 'pink', linewidth=7) 
plt.show    ## linestyle- dashed/dotted , linewidth ,
```

```python
x=np.array([100,300,450,890])
y=np.array([240,340,650,390])

plt.plot(x,y,color= 'pink' ,linewidth = 2)

plt.xlabel("Average pulse",fontsize=10,color='red')
plt.ylabel("Calorie burner",color='red')
plt.title("Sample",fontsize=30,color='k')
plt.show()
```

**WITH PYPLOT, YOU CAN USE THE title() FUNCTION TO SET A TITLE FOR THE PLOT**



```python
x=np.array([100,300,450,890])
y=np.array([240,340,650,390])
 
##adding the label location with (loc='left/right/top/bottom')

plt.title("Production Rate",fontsize=35,loc='left')
plt.xlabel("Duration",fontsize=20,color='r',loc='right')
plt.ylabel("Quality",fontsize=20,color='r',loc='top')

plt.plot(x,y)

plt.show()
```

```python
### ADDING GRID AND ITS COLOR ###

x=np.array([100,300,450,890])
y=np.array([240,340,650,390])

plt.title("Production Rate",fontsize=35,loc='left')
plt.xlabel("Duration",fontsize=20,color="green",loc='right')
plt.ylabel("Quality",fontsize=20,color='r',loc='top')

plt.plot(x,y,color='green')

plt.grid(color='orange',linestyle='dotted')
plt.show()
```

```python
### YOU CAN USE THE AXIS PARAMETER IN grid() function TO SPECIFY WHICH GRID LONES TO SHOW #.

# LEGAL  VALUE ARE 'x' 'y' and 'both'. DEFAULT IS 'both'

x=np.array([100,300,450,890])
y=np.array([240,340,650,390])

plt.title("Production Rate",fontsize=35,loc='left')
plt.xlabel("Duration",fontsize=20,color="green",loc='right')
plt.ylabel("Quality",fontsize=20,color='r',loc='top')

plt.plot(x,y,color='yellow')   # CAN BE DONE WITH 'y'

plt.grid(axis='x')
plt.show()
```

```python
# TO ADD THE GRID IN BOTH THE AXIS THE VALUE SHOULD BE EQUAL X=Y

x=np.array([100,300,450,890])
y=np.array([240,340,650,390])

plt.title("Production Rate",fontsize=35,loc='left')
plt.xlabel("Duration",fontsize=20,color="green",loc='right')
plt.ylabel("Quality",fontsize=20,color='r',loc='top')

plt.plot(x,y,color='green')
plt.grid(color='yellow',linestyle='dotted',linewidth=2)  # supported values are '-','--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.show()
```
